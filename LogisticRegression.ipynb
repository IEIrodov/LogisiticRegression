```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
```

```python
cancer_ds = datasets.load_breast_cancer()
cancer_ds
```

```python
clf = LogisticRegression(max_iter=10000,C=2,multi_class='multinomial',solver='lbfgs') 
clf.fit(cancer_ds.data,cancer_ds.target)

#multi_class is used to pick the techniques in multiclass LogisticRegression like we want one verus all or Multinomial Logistic Regression
#Solver:- Which Algorithm we are using currently we are using 'lbfgs' which is kind of slower algorithm if we want faster results 
# We should use 'sag' or 'saga'


#Here C comes with the actual cost function which means huge 'C' will lead to Overfitting of data
# Because the value of lambda(sum(m^2)) is neglected due to high value of C and it's vice versa is also true
```

```python
#Accuracy of Model
#We see a change in Score or our Accuracy is getting better when we use multi_class='multinomial' approach
# instead of one versus many which we studied in theory also

#Accuracy wrt to algorithm:- sag = 92% = saga,lbfgs = 96%
clf.score(cancer_ds.data,cancer_ds.target)
```

```python
#Prob for each test point
clf.predict_proba(cancer_ds.data)[13]
```
