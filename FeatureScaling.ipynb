```python
import numpy as np
from sklearn import preprocessing
```

```python
X = np.array([[1.,-1.,2.],
              [2.,0.,0.],
              [0.,1.,-1.]])
```

```python
x_scaled = preprocessing.scale(X) #Scale basically tries to make mean to be 0 and variance to be 1
```

```python
x_scaled.mean(axis=0) #Mean around x
```

```python
x_scaled.std(axis=0) #Standard Deviation = 1
```

```python
scaler = preprocessing.StandardScaler()
scaler.fit(X)
scaler.transform(X) # Same as scale but here we have store the parameters for future scaling of our testing data
```

```python
X_test = [[1,1,0]]
scaler.transform(X_test)
```
